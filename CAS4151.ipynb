```python
from fastapi import FastAPI
from fastapi import Body, Form
from fastapi.middleware.cors import CORSMiddleware
import json

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)

@app.get('/hello')
async def hello(name : str) :
    return {'response' : 'hello, '+name}

@app.post('/echo')
async def echo(request: str = Form(...)):
    data = json.loads(request)
    return data

@app.post('/data')
async def echo(data: str = Form(...), request: str = Form(...)):
    json_data = json.loads(request)
    ret_data = {"json" : json_data['data'], "string": data}
    return ret_data
```
